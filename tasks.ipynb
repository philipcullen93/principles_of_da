## Task 6: Analyse Relationships

Using numpy.polyfit we can add a linear regression line to our plot from task 5.

```python
# Creation of scatter plot by class
plt.scatter(petal_l[target==0], petal_w[target==0], color = 'red', label = target_names[0], alpha = 1)
plt.scatter(petal_l[target==1], petal_w[target==1], color = 'blue', label = target_names[1], alpha = 1)
plt.scatter(petal_l[target==2], petal_w[target==2], color = 'green', label = target_names[2], alpha = 1)

# Linear rgression for all data
x = petal_l
y = petal_w
slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept

# Plot regression line
plt.plot(x, regression_line, color='black', label='Linear Regression')

# Labelling of scatter plot
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs. Petal Width')
plt.grid(True)
plt.legend()
plt.show()
```

## Task 9: Fit a Simple Linear Regression

```python
# Linear regression using numpy
coeffs = np.polyfit(petal_l, petal_w, 1)
slope, intercept = coeffs
y_pred = slope * petal_l + intercept

# R^2 calculation
ss_res = np.sum((petal_w - y_pred) ** 2)
ss_tot = np.sum((petal_w - np.mean(petal_w)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Plot
plt.scatter(petal_l[target==0], petal_w[target==0], color='red', label=target_names[0])
plt.scatter(petal_l[target==1], petal_w[target==1], color='blue', label=target_names[1])
plt.scatter(petal_l[target==2], petal_w[target==2], color='green', label=target_names[2])

x_vals = np.linspace(min(petal_l), max(petal_l), 100)
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, color='black', label='Linear Fit')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs. Petal Width')
plt.text(1, 2.2, f'$R^2 = {r2:.2f}$', fontsize=12, bbox=dict(facecolor='white'))
plt.grid(True)
plt.legend()
plt.show()
```

It is also worth noting that while the above linear regression and R^2 values were determined and plotted using numpy. Modules such as scipy and scikit-learn could also be used instead.

## Task 10: Too Many Features

Pairplots are extremely useful when looking at data as it creates a grid of scatter plots and histograms.
In this case, our pairplot will create scatter plots and histograms comparing the features of the Iris Dataset.
This gives us an overview of how the features relate to classification, helps visualise the data between multiple variables at once, and suggests which features best separate species.

```python
# load data using seaborn
iris_sbrn = sbrn.load_dataset("iris")

# Define custom color palette.
custom_palette = {
    "setosa": "red",
    "versicolor": "blue",
    "virginica": "green"
    }

# Create a pairplot
sbrn.pairplot(iris_sbrn, hue="species", palette=custom_palette)
plt.show()
```

As I wanted the plots to resemble ones I produced earlier I added a line of code to change the colours to red for setosa, blue for veriscolor, and green for virginica.

The resulting pairplot provides a 4 x 4 grid, with histograms representing the comparison between the different features being seen on the diagonal, while scatter plots comparing the different features make are shown off the diagonal.

By looking at the resulting plots we can draw a number of conclusions about the relationship between features.
- Petal length vs petal width shows a strong positive correlation across all species.
- While Sepal length vs sepal width shows a weaker or no consistent correlation.

We can also assess with features best separate the individual species.
- Setosa is clearly separated from the others in most plots involving petal features. In general it appears to cluster away from both  Versicolor and Virginica.
- Versicolor and Virginica tend to overlap more, but still show some distinction - particularly when looking at petal lenght, and petal width.

We can draw an overall conclusion about several features.
- Setosa is easily distinguishable from the other two species based on petal size.
- Versicolor and Virginica overlap, but by looking at petal length/width it is more easy to seaparte them separate them.
 - Petal features appear to be more useful for classification and clustering than sepal features.
