## Task 3: Summarize the Data

For each feature in the dataset, calculate and display:

We have 4 features so we can use numpy to calculate the mean, minimum, maximum, standard deviation and median for each feature. Numpy is imported as np to make coding more effecient. I also convert our the data into a numpy array using the np.array() function. This allows for more efficient use of operations on the data.

Calculations (cm)

```python
# convert the data to a numpy array
np_data = np.array(data)
```

```python
# column_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
column_names = iris.feature_names
```

The code below reads each column and applies the np function as below. It then moves onto the next column. Since we know the feature names we can have the code apply the appropriate name to each calcumation.

```python
# this code is a loop that loops through each column in the dataset
# it calculates the statistical data for each column
# the statisitical data calculated is: mean, minimum, maximum, standard deviation, and median

# Loops through each column in the dataset, i = each column in the dataset
for i in range(data.shape[1]):

    # extracts column i
    column = data[ :, i]
    # prints the name of each column
    print(f"{column_names[i]}:")
    # calculates and prints each of the statistical data
    print("Mean:", float(np.mean(column)))
    print("Minimum:", float(np.min(column)))
    print("Maximum:", float(np.max(column)))
    print("Standard Deviation:", float(np.std(column)))
    print("Median:", float(np.median(column)))
    print()
```

## Task 6: Analyse Relationships

Using numpy.polyfit we can add a linear regression line to our plot from task 5.

```python
# Creation of scatter plot by class
plt.scatter(petal_l[target==0], petal_w[target==0], color = 'red', label = target_names[0], alpha = 1)
plt.scatter(petal_l[target==1], petal_w[target==1], color = 'blue', label = target_names[1], alpha = 1)
plt.scatter(petal_l[target==2], petal_w[target==2], color = 'green', label = target_names[2], alpha = 1)

# Linear rgression for all data
x = petal_l
y = petal_w
slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept

# Plot regression line
plt.plot(x, regression_line, color='black', label='Linear Regression')

# Labelling of scatter plot
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs. Petal Width')
plt.grid(True)
plt.legend()
plt.show()
```

## Task 9: Fit a Simple Linear Regression

```python
# Linear regression using numpy
coeffs = np.polyfit(petal_l, petal_w, 1)
slope, intercept = coeffs
y_pred = slope * petal_l + intercept

# R^2 calculation
ss_res = np.sum((petal_w - y_pred) ** 2)
ss_tot = np.sum((petal_w - np.mean(petal_w)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Plot
plt.scatter(petal_l[target==0], petal_w[target==0], color='red', label=target_names[0])
plt.scatter(petal_l[target==1], petal_w[target==1], color='blue', label=target_names[1])
plt.scatter(petal_l[target==2], petal_w[target==2], color='green', label=target_names[2])

x_vals = np.linspace(min(petal_l), max(petal_l), 100)
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, color='black', label='Linear Fit')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs. Petal Width')
plt.text(1, 2.2, f'$R^2 = {r2:.2f}$', fontsize=12, bbox=dict(facecolor='white'))
plt.grid(True)
plt.legend()
plt.show()
```

It is also worth noting that while the above linear regression and R^2 values were determined and plotted using numpy. Modules such as scipy and scikit-learn could also be used instead.
